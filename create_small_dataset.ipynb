```python
import pandas as pd

df = pd.read_csv('preprocessed_movies.csv',index_col='ID')
df.columns
```

```python
allgenres = df.columns.tolist()[3:]
subsetgenres = ['Action','Drama','Animation']
genrestodrop = [g for g in allgenres if g not in subsetgenres]
print(allgenres)
print(genrestodrop)
```

```python
smalldf = df[(df['Action'] == 1) | (df['Drama'] == 1) | (df['Animation'] == 1)]
```

```python
smalldf.drop(genrestodrop, axis=1, inplace=True)
smalldf.index.name = 'ID'
smalldf.sample(50)
```

```python
smalldf.to_csv('smalldf.csv')
```
